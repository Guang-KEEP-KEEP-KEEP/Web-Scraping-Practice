```python
from DrissionPage import ChromiumPage
import time

# 初始化页面
page = ChromiumPage()


# 目标关键词
keyword = '穿搭'
url = f'https://www.xiaohongshu.com/search_result?keyword={keyword}'
page.get(url)
```

```python
prev_count = 0
note_ids = set()
unique_notes = set()  # 用来存储去重后的 title 和 author 信息
success = 0  # 计数成功提取的笔记内容

for i in range(30):
    page.run_js("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(random.uniform(4, 6))  # 给加载更多时间

    cards = page.eles('css:section.note-item')
    print(f'滚动第 {i+1} 次，卡片数：{len(cards)}')

    for card in cards:
        a_tag = card.ele('tag:a', timeout=0.5)
        if a_tag:
            href = a_tag.attr('href')
            if href and '/explore/' in href:
                note_id = href.split('/explore/')[-1]
                note_ids.add(note_id)#这个部分的目的只是为了提取card id，其实非必要

                # 尝试提取 title 和 author
                try:
                    title = card.ele('css:a.title', index=0, timeout=1).text
                    author = card.ele('css:span.name', index=0, timeout=1).text
                    # 使用元组 (title, author) 来去重
                    if (title, author) not in unique_notes:
                        unique_notes.add((title, author))
                        print(f'{success+1}: {title} | {author}')
                        success += 1
                except Exception as e:
                    print(f'{success+1}: 解析失败 -> {e}')

    # 判断有没有加载新内容
    if len(note_ids) == prev_count:
        print("📌 没有加载新内容，提前停止")
        break
    else:
        prev_count = len(note_ids)

print(f'\n✅ 共成功提取 {success} 条笔记内容')
print(f'✅ 共获取到 {len(note_ids)} 条唯一笔记')
```
