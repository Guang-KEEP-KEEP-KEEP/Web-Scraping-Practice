```python
from googleapiclient.discovery import build

# 设置 API 密钥
api_key = 'AIzaSyAYnykRSTE0GZyYUw2MsyD82l6yynZ2r1M'  # 请替换为你的 API 密钥

# 初始化 YouTube API 客户端
youtube = build("youtube", "v3", developerKey=api_key)

# 获取频道的所有视频 ID
def get_video_ids_from_channel(channel_id, max_results=50):
    video_ids = []
    next_page_token = None

    while True:
        # 通过搜索 API 获取频道的所有视频
        request = youtube.search().list(
            part="id",
            channelId=channel_id,
            maxResults=max_results,
            pageToken=next_page_token
        )
        response = request.execute()

        # 收集视频 ID
        for item in response['items']:
            if item['id']['kind'] == 'youtube#video':  # 确保是视频，而不是播放列表或其他类型
                video_ids.append(item['id']['videoId'])

        # 检查是否有下一页
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return video_ids

# 获取视频的详细信息（如标题、观看量等）
def get_video_details(video_ids):
    video_details = []
    
    for i in range(0, len(video_ids), 50):  # YouTube API 每次最多请求 50 个视频
        video_ids_chunk = video_ids[i:i+50]
        
        request = youtube.videos().list(
            part="snippet,statistics",
            id=','.join(video_ids_chunk)
        )
        response = request.execute()

        # 提取视频信息
        for item in response['items']:
            video_info = {
                'video_id': item['id'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'publish_date': item['snippet']['publishedAt'],
                'view_count': item['statistics'].get('viewCount', 'N/A'),
                'like_count': item['statistics'].get('likeCount', 'N/A'),
                'comment_count': item['statistics'].get('commentCount', 'N/A')
            }
            video_details.append(video_info)
    
    return video_details

# 示例：使用频道 ID 获取视频数据
channel_id = 'UCBa5G_ESCn8Yd4vw5U-gIcg'  

# 步骤 1：获取视频 ID 列表
video_ids = get_video_ids_from_channel(channel_id)

# 步骤 2：获取每个视频的详细信息
video_data = get_video_details(video_ids)

# 输出视频信息
for video in video_data:
    print(f"视频标题: {video['title']}")
    print(f"发布日期: {video['publish_date']}")
    print(f"观看次数: {video['view_count']}")
    print(f"点赞次数: {video['like_count']}")
    print(f"评论次数: {video['comment_count']}")
    print(f"视频链接: https://www.youtube.com/watch?v={video['video_id']}")
    print("=" * 50)
```
